```python
#%%
import pandas as pd
import numpy as np
from tqdm.notebook import trange, tqdm
```

```python
# Layout of the columns
layout_df = pd.read_csv('/Users/thomas.farrandibm.com/Documents/GitHub/vietnam_air_records/data/HERB/RG330.HERBO2.ORIG',
                        delimiter='\t',
                        skipinitialspace=True,
                        header=None)
layout_df
```

```python
def string_slicer(row, document_start_char, document_end_char):
    start_char = document_start_char - 1
    end_char = document_end_char - 1
    
    if start_char == end_char:
        sliced_string = row[start_char]
    else:
        sliced_string = row[start_char:end_char+1]
        
    return sliced_string
```

```python
new_list = []

with tqdm(total=len(layout_df[0])) as pbar:
    for row in layout_df[0]:
        row_entry = {
            'Date of Flight': string_slicer(row, 1, 6),
            'Combat Tactical Zone': string_slicer(row, 7, 7),
            'Mission Project Number': string_slicer(row, 8, 15),
            'Multiple Run of Mission': string_slicer(row, 16, 16),
            'Multiple Province Mission Coverage Indicator': string_slicer(row, 17, 17),
            'Blank': string_slicer(row, 18, 18),
            'Province Code': string_slicer(row, 19, 20),
            'Number of Aircraft Scheduled': string_slicer(row, 21, 22),
            'Number of Airborne Aircraft': string_slicer(row, 23, 24),
            'Number of Productive Aircraft': string_slicer(row, 25, 26),
            'Type of Agent': string_slicer(row, 27, 28),
            'Number of Gallons': string_slicer(row, 29, 33),
            'Number of Hits': string_slicer(row, 34, 35),
            'Number of Maintenance Aborts': string_slicer(row, 36, 36),
            'Number of Weather Aborts': string_slicer(row, 37, 37),
            'Number of Battle Damage Aborts': string_slicer(row, 38, 38),
            'Other Types of Aborts': string_slicer(row, 39, 39),
            'Type of Mission': string_slicer(row, 40, 40),
            'Blank': string_slicer(row, 41, 48),
            'Area Sprayed': string_slicer(row, 49, 52),
            'Result (Helo Only)': string_slicer(row, 53, 53),
            'Blank': string_slicer(row, 54, 86),
            'Mission Leg Control Field': string_slicer(row, 87, 88),
            'Blank': string_slicer(row, 89, 97),
            'UTM': string_slicer(row, 98, 105),
        }
        new_list.append(row_entry)
        
df = pd.DataFrame(new_list)
df.drop('Blank', axis=1, inplace=True)
df
```
