```python
#%%
import pandas as pd
import numpy as np
from tqdm.notebook import trange, tqdm
import folium
from folium import plugins
```

```python
# Layout of the columns
layout_df = pd.read_csv('/Users/thomas.farrandibm.com/Documents/GitHub/vietnam_air_records/data/SITRA/SITRA.TR.6668',
                        delimiter='\t',
                        skipinitialspace=True,
                        header=None)
layout_df
```

```python
def string_slicer(row, document_start_char, document_end_char):
    start_char = document_start_char - 1
    end_char = document_end_char - 1
    
    if start_char == end_char:
        sliced_string = row[start_char]
    else:
        sliced_string = row[start_char:end_char+1]
        
    return sliced_string
```

```python
new_list = []

with tqdm(total=len(layout_df[0])) as pbar:
    for row in layout_df[0]:
        row_entry = {
            'DATE: Message Date': string_slicer(row, 1, 6),
            'DATAT: Data Type': string_slicer(row, 7, 7),
            'SERAL: Paragraph Number': string_slicer(row, 8, 10),
            'CORP: Military Region': string_slicer(row, 11, 11),
            'PART: Part': string_slicer(row, 12, 12),
            'FNATL: Force Nationality': string_slicer(row, 13, 13),
            'CNTRY: Country Code': string_slicer(row, 14, 15),
            'OPNAM: Operation Name': string_slicer(row, 16, 40),
            'OPCOD: Operation Type Code': string_slicer(row, 41, 42),
            'UDATE: Date of Last Update': string_slicer(row, 43, 47),
            'PRVC1: Major Province Code': string_slicer(row, 48, 49),
            'PRVC2: Secondary Province Code': string_slicer(row, 50, 51),
            'PRVC3: Tertiary Province Code': string_slicer(row, 52, 53),
            'OPDAY: Battalion Days of Operation': string_slicer(row, 54, 57),
            'CTDAY: Number of Days of Contact': string_slicer(row, 58, 58),
            'DTEI: Initiation Date': string_slicer(row, 59, 64),
            'HOURI: Initation Hour': string_slicer(row, 65, 68),
            'DTET: Termination Date': string_slicer(row, 69, 74),
            'HOURT: Termination Hour': string_slicer(row, 75, 78),
            'FSLIN: Line Number of First Set': string_slicer(row, 79, 81)
        }
        new_list.append(row_entry)
        
df = pd.DataFrame(new_list)
df
```
